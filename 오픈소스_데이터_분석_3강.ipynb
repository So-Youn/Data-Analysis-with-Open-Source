## 3-1 언패킹

```python
# 튜플 : 순서가 있는 변경 불가능한 데이터 나
coordinates = (10, 20)
x,y = coordinates
# 튜플 언패킹

print(f'x좌표: {x}, y좌표: {y}')

rgb = [255, 128, 0]
# 리스트 언패킹
red,green,blue = rgb
print(f'빨강: {red}, 초록: {green}, 파랑: {blue}')

word = 'ABC'
# 문자열 언패킹
first, second, third = word
print(f'첫 번째 문자: {first}, 두 번째 문자: {second}, 세 번째 문자: {third}')
```

```python
def get_dataset_stats(values):
  return min(values), max(values), sum(values) / len(values)

data = [12, 8, 21, 17, 5, 32, 14]

# 함수값 return 언패킹
minimum, maximum, average = get_dataset_stats(data)
print(f'최솟값: {minimum}, 최댓값: {maximum}, 평균: {average:.2f}')
```

```python
student_scores = [
    ('김철수', 85, 92, 78),
    ('이영희', 92, 88, 95),
    ('박지민', 75, 83, 90)
]

# 반복문에서 언패킹 사용
for name, database, python, cloud in student_scores:
  average = (database + python + cloud) / 3
  print(f'{name}의 평균 점수: {average:.1f}')
```

## 3-3 함수형 프로그래밍

```python
# lambda: 익명함수, 간단한 함수를 한 줄로 만들 때 사용 (한 번만, 즉석에서 쓸 )
# lambda 매개변수: 표현식
employees = {
    '이지혜': 3000000,
    '구민준': 5000000,
    '방서연': 4100000
}

# 함수 정의: salary * 1.1
adjust_salary = lambda salary: salary * 1.1
# 딕셔너리 컴프리헨션
updated_salaries = {name: adjust_salary(salary) for name, salary in employees.items()}
# 인상된 급여 출력
print(updated_salaries)
```

**map함수**
 - 반복 가능한 객체의 모든 요소에 특정 함수를 적용하여 새로운 값을 생성하는 함수
 - 리스트나 튜플과 같은 자료구조의 요소를 변경할 때 유용

```python
temperature_celsius = [25.6, 27.8, 30.5, 22.3, 28.9, 31.2, 24.7]

# 함수 정의: 섭씨 -> 화씨
cel_to_fah = lambda c: (c*9/5) + 32

# 온도 변환
# map(함수, 반복 가능한 객체(ex, 리스트, 튜플, 레인지 ))
## temperature_celsius 리스트의 각 요소(25.6, 27.8, 30.5, ...)를 하나씩 꺼내서 cel_to_fah 함수에 전달
## map의 결과는 Map객체(iterator)이므로 보통 list()로 감싸서 리스트로 변환한다.
temperature_fahrenheit = list(map(cel_to_fah, temperature_celsius))

print('섭씨 온도 데이터:', temperature_celsius)
print('화씨 온도로 변환:', temperature_fahrenheit)
```

```python
# 딕셔너리
employees = [
    {'name': '김지원', 'salary': 3700000},
    {'name': '박민준', 'salary': 7000000},
    {'name': '이서연', 'salary': 4400000},
    {'name': '정현우', 'salary': 3500000}
]

# salary * 1.1
# employees의 각 요소(딕셔너리)를 하나씩 가져다가 lambda 함수에 전달
## map 함수 updated_employees = list(map(..., employees))
## {'name': e['name'], 'salary': e['salary'] * 1.1} : 새로운 딕셔너리 생성
updated_employees = list(map(lambda e: {'name': e['name'], 'salary': e['salary'] * 1.1}, employees))

print('\n급여 인상 후:')
for emp in updated_employees:
    print(f"{emp['name']}: {int(emp['salary']):,}원")
```

**filter**
- 반복 가능한 객체의 요소 중 특정 조건을 만족하는 요소만 남기는 함수
- 조건 함수가 True를 반환하는 요소만 컬렉션에 포함

```python
numbers = [10, 15, 20, 25, 30]

# 짝수 필터링
# filter() 함수는 특정 조건을 만족하는 요소만 남기고 싶을 때 사용하는 파이썬 내장 함수
# filter(람다 함수, numbers)  <- filter(함수, iterable)
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
# 리스트 컴프리헨션 사용
even_numbers_comp = [x for x in numbers if x % 2 == 0]

print(even_numbers)
print(even_numbers_comp)
```

```python
osda_students = [
    {'name': '김지원', 'score': 85},
    {'name': '이민준', 'score': 65},
    {'name': '박서연', 'score': 90},
    {'name': '정현우', 'score': 55},
    {'name': '최예은', 'score': 78}
]

# 70점 이상 학생 목록 생성
passed_students = list(filter (lambda student: student['score'] >= 70, osda_students))

for student in passed_students:
    print(f'이름: {student["name"]}, 점수: {student["score"]}')
```

**reduce**
- 반복 가능한 객체의 모든 요소를 누적하여 하나의 값으로 축소하는 함수
- 반복문을 사용하지 않고도 데이터를 하나의 값으로 집계
- functools 모듈에서 제공

```python
# reduce()는 리스트 전체를 하나의 값으로 축약(reduce) 하는 함수
# reduce(함수, iterable): 함수는 반드시 두 개의 인자 가져야 함.
# 그 결과, reduce()는 요소를 두 개씩 줄여나가면서 최종적으로 하나의 값만 남긴다.
from functools import reduce

numbers = [3, 7, 2, 9, 5]
# 최대값 생성
max_value = reduce(lambda x, y: x if x > y else y, numbers) # x가 크면 x, 아니면 y
print(max_value)
```

```python
from functools import reduce
numbers = [1,2,3,4,5]
total_sum = reduce(lambda x,y: x+y, numbers) # 첫번째와 두번째 요소를 함수에 적용하여 중간 결과를 생성
print(total_sum)
```
